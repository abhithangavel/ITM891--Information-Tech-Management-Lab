### Problem 1 - Basic Features of Graphs
Here we will create a complete graph of 5 vertices and determine basic features about the graph, such as the number of edges it has, the degree, etc. Refer to the documentation here for help: https://networkx.org/documentation/stable/tutorial.html.



```python
import networkx as nx

# Enter your answer under each step.

# Create a complete graph of 5 vertices
G1 = nx.complete_graph(5)

# Plot the graph
import matplotlib.pyplot as plt
nx.draw(G1, with_labels=True)
plt.show()
# How many edges in this graph?
print("Number of edges:", G1.number_of_edges())

# What is the adjacency matrix for this graph?
adj_matrix = nx.adjacency_matrix(G1)
print("Adjacency matrix:\n", adj_matrix.todense())

# What is the degree for every vertex?
degree_dict = dict(G1.degree())
print("Degrees:", degree_dict)

# What are the neighbors of the vertex 0 in this graph?
print("Neighbors of vertex 0:", list(G1.neighbors(0)))
```

```python
# Calculate the shortest path between any two vertices (all the pairs in the graph).
for source in G1.nodes:
    for target in G1.nodes:
        if source != target:
            shortest_path = nx.shortest_path(G1, source=source, target=target)
            print(f"Shortest path between nodes {source} and {target}: {shortest_path}")
```

```python
# Calculate the distance between each pair of vertices (0 and 1, 1 and 2, 2 and 3, etc.).
for i in range(G1.number_of_nodes()):
    for j in range(i+1, G1.number_of_nodes()):
      distance_matrix = nx.shortest_path_length(G1, i, j)
      print(f"Distance between {i} and {j} is {distance_matrix}")
```

```python
# Calculate the diameter of the graph.
diameter = nx.diameter(G1)
print(f"The diameter of the karate club graph is {diameter}.")
```

```python
# Is the graph connected? Check to see if it is. 
print("Is the graph connected?", nx.is_connected(G1))
```

```python
# How many connected components are in the graph?
print("Number of connected components:", nx.number_connected_components(G1))
```

###Problem 3 - Karate Club Graph

```python
# There are some existing graphs that you can load in the networkx.
# Let's play with karate_club graph. Create an instance of this graph.
G3 = nx.karate_club_graph()

# Draw it
nx.draw(G3, with_labels=True)
plt.show()
```

```python
# How many connected components are in this graph?
num_components = nx.number_connected_components(G3)

print(f"The karate club graph has {num_components} connected component(s).")
```

```python
# Print the degree of each vertex in this graph
Deg = G3.degree()
print("Degree of each vertex is: ")
for node, degree in Deg:
  print("Node {}: Degree {}".format(node, degree))
print(" ")
```

```python
# What is the diameter of this graph?
diameter = nx.diameter(G3)
print(f"The diameter of the karate club graph is {diameter}.")
```

```python
# What is the path with the distance equal to this diameter?
# Hint: Longest distance, nx.shortest_path(G)

# Find the pair of nodes with the longest distance
max_distance = -1
max_pair = None
# Add more code here.... 
list(nx.shortest_path_length(G3))
for source_node, distances in nx.shortest_path_length(G3):
  for target_node, distance in distances.items():
    if distance > max_distance:
      max_distance = distance
      max_pair = (source_node, target_node)
print(f"Max distance: {max_distance}")
print(f"The pair of nodes with the longest distance in the karate club graph is {max_pair}.")

# Get the path corresponding to the longest distance
path = nx.shortest_path(G3, source=max_pair[0], target=max_pair[1])
print(f"The path corresponding to the longest distance is {path}.")
```
