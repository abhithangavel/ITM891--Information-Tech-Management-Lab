## More about clustering....



###Problem 1###
There are many other clustering algorithms implemented in the scikit learn library. Let's dive into them one by one.

```python
# First, let's take a look at the blob data. Here we make use of sklearn.datasets.make_blobs.

############## Given code ##############
from sklearn.datasets import make_blobs
X,y = make_blobs(n_samples=100, n_features=2, random_state=1, centers=3)
########################################

# Create a scatterplot from the data X with different colors according to y values. 
# There should be three different clusters.

########## Your answer here ############
import matplotlib.pyplot as plt

plt.scatter(X[:,0],X[:,1], c = y)
########################################
```

###Problem 2

Now, let's apply different clustering algorithms. 

```python
# Let's use different modules from sklearn.cluster and visualize your clusters.
# Here we are using KMeans. You will need to use this to predict and plot the data using a scatter plot. 

############## Given code ##############
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42,n_init="auto")
########################################

########## Your answer here ############
import numpy as np

kmeans.fit(X)
colors = ['r','g','b']

cluster_pred = kmeans.predict(X)
for i in range(3):
  plt.scatter(X[cluster_pred==i][:,0],X[cluster_pred==i][:,1],color = colors[i])
########################################
```

```python
# Here we are using SpectralClustering. You will need to use this to predict and plot the data using a scatter plot. 

############## Given code ##############
from sklearn.cluster import SpectralClustering
spec = SpectralClustering(n_clusters=3)
########################################

########## Your answer here ############
cluster_labels = spec.fit_predict(X)
colors = ['r','g','b']

for i in range(3):
  plt.scatter(X[cluster_labels==i][:,0],X[cluster_labels==i][:,1],color = colors[i])
########################################
```

```python
# Here we are using GaussianMixture. You will need to use this to predict and plot the data using a scatter plot. 

############## Given code ##############
from sklearn.mixture import GaussianMixture
GM = GaussianMixture(n_components=3, random_state=42)
########################################

########## Your answer here ############
GM.fit(X)
cluster_labels = GM.predict(X)
colors = ['r','g','b']

for i in range(3):
  plt.scatter(X[cluster_labels==i][:,0],X[cluster_labels==i][:,1],color = colors[i])

########################################
```

```python
# Here we are using OPTICS. You will need to use this to predict and plot the data using a scatter plot. 

############## Given code ##############
from sklearn.cluster import OPTICS
optics = OPTICS(min_samples=20)
########################################

########## Your answer here ############
cluster_labels = optics.fit_predict(X)
colors = ['r','g','b']

for i in range(3):
  plt.scatter(X[cluster_labels==i][:,0],X[cluster_labels==i][:,1],color = colors[i])
########################################
```

###Problem 3

Let's try different datasets and reapply all these different clustering methods including KMeans, SpectralClustering, GaussianMixture, and OPTICS.

```python
# We are going to use a different data set to see how the clustering works with this new data. 
# Take a look at this link to understand what the output will look like. 
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_circles.html
# Do similar to what you did above in Problem 2 for each part. 

############## Given code ##############
from sklearn import datasets
X,y = datasets.make_circles(200)
########################################
colors = ['r','g']

########## Your answer here ############
# KMeans...
kmeans = KMeans(n_clusters=2, random_state=42,n_init="auto")

kmeans.fit(X)

cluster_pred = kmeans.predict(X)
for i in range(2):
  plt.scatter(X[cluster_pred==i][:,0],X[cluster_pred==i][:,1],color = colors[i])
plt.show()
########################################


########## Your answer here ############
# SpectralClustering... 
spec = SpectralClustering(n_clusters=2)
cluster_labels = spec.fit_predict(X)

for i in range(2):
  plt.scatter(X[cluster_labels==i][:,0],X[cluster_labels==i][:,1],color = colors[i])
plt.show()
########################################


########## Your answer here ############
# GaussianMixture...
GM = GaussianMixture(n_components=2, random_state=42)

GM.fit(X)
cluster_labels = GM.predict(X)

for i in range(2):
  plt.scatter(X[cluster_labels==i][:,0],X[cluster_labels==i][:,1],color = colors[i])
plt.show()
########################################

########## Your answer here ############
# OPTICS...
optics = OPTICS(min_samples=5)
cluster_labels = optics.fit_predict(X)

for i in range(2):
  plt.scatter(X[cluster_labels==i][:,0],X[cluster_labels==i][:,1],color = colors[i])
plt.show()
########################################
```

```python
# Reapply all these different clustering methods to another new dataset. 
# Here we are using make_blobs. Refer to this link: 
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html

############## Given code ##############
from sklearn import datasets
import numpy as np
X, y = datasets.make_blobs(n_samples=100, random_state=3)
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X = np.dot(X, transformation)
########################################
colors = ['r','g','b']

########## Your answer here ############
# KMeans...
kmeans = KMeans(n_clusters=3, random_state=42,n_init="auto")

kmeans.fit(X)

cluster_pred = kmeans.predict(X)
for i in range(3):
  plt.scatter(X[cluster_pred==i][:,0],X[cluster_pred==i][:,1],color = colors[i])
plt.show()
########################################


########## Your answer here ############
# SpectralClustering... 
spec = SpectralClustering(n_clusters=3)
cluster_labels = spec.fit_predict(X)

for i in range(3):
  plt.scatter(X[cluster_labels==i][:,0],X[cluster_labels==i][:,1],color = colors[i])
plt.show()
########################################


########## Your answer here ############
# GaussianMixture...
GM = GaussianMixture(n_components=3, random_state=42)

GM.fit(X)
cluster_labels = GM.predict(X)

for i in range(3):
  plt.scatter(X[cluster_labels==i][:,0],X[cluster_labels==i][:,1],color = colors[i])
plt.show()
########################################

########## Your answer here ############
# OPTICS...
optics = OPTICS(min_samples=15)
cluster_labels = optics.fit_predict(X)

for i in range(3):
  plt.scatter(X[cluster_labels==i][:,0],X[cluster_labels==i][:,1],color = colors[i])
plt.show()
########################################
```

```python
############## Given code ##############
X,y = datasets.make_circles(200)
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, linkage='ward') # change the min_samples to see different clustering results
########################################

########## Your answer here ############
cluster_labels = cluster.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels)
plt.show()
########################################

############## Given code ##############
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, linkage='complete') # change the min_samples to see different clustering results
########################################

########## Your answer here ############
cluster_labels = cluster.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels)
plt.show()
########################################

############## Given code ##############
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, linkage='average') # change the min_samples to see different clustering results
########################################

########## Your answer here ############
cluster_labels = cluster.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels)
plt.show()
########################################

############## Given code ##############
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, linkage='single') # change the min_samples to see different clustering results
########################################

########## Your answer here ############
cluster_labels = cluster.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels)
plt.show()
########################################
```
